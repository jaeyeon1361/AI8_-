```python
raw_inputs =["나는 학생입니다.",
             "나는 좋은 선생님 입니다.",
             "당신은 매우 좋은 선생님 입니다."]
raw_labels=[1,0,0]
```

```python
words = []
for s in raw_inputs : 
  words.extend(s.split())
print(words)
```

```python
# 중복 단어 제거
words = list(dict.fromkeys(words))
print(words)
```

```python
word_to_id = {"<PAD>" : 0 , "<UNK" : 1}

for w in words : 
  word_to_id[w] = len(word_to_id) # 단어를 하나하니씩 추가
  print(word_to_id)
```

```python
word_to_id.items() # key와 values 값을 한꺼번에 출력
```

```python
id_to_words = {i : w  for w, i in word_to_id.items()}
print(id_to_words)
```

```python
train_inputs = [] # 문장마다 시퀀스 변환하는 과정


for s in raw_inputs : 
  row = [word_to_id[w] for w in s.split()]

  # padding
  row += [0] * (5-len(row))
  train_inputs.append(row)

train_inputs = np.array(train_inputs)
print(train_inputs)
```

```python
onehot_matrix = np.eye(len(word_to_id))
print(onehot_matrix)
```

```python
train_onehot = onehot_matrix[train_inputs]
print(train_onehot)
```

```python
train_onehot.shape
```

```python
print(np.argmax(train_onehot, axis = -1))
```

```python
x = np.argmax(train_onehot, axis = -1)
print(x)
```

```python
import tensorflow.keras.layers as L
```

```python
x_len = train_onehot.shape
print(x_len)
```

```python
inp = tf.convert_to_tensor(x, dtype=tf.int32) # 정수 시퀀스
inp_len = tf.convert_to_tensor(x_len, dtype=tf.int32)
```
