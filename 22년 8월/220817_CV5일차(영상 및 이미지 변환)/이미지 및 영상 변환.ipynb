```python
import cv2
import numpy as np
import sys
import matplotlib.pyplot as plt
```

```python
import glob
```

```python
img_files = glob.glob('./fig/my_images/*.*')
idx=0
cv2.namedWindow('scene', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

while True : 
    img = cv2.imread(img_files[idx])
    
    if img is None :
        print('image read failed')
        break
        
    cv2.imshow('scene', img)
    
    if cv2.waitKey(3000) == 27:
        break
        
    idx += 1
    
    if idx >= len(img_files) :
        idx = 0
        
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened() : 
    print('camera open failed')
    sys.exit()
    
while True :
            
    ret, frame = cap.read()
    
    if not ret : 
        print('frame open failed')
        break
        
    cv2.imshow('image', frame)
    
    if cv2.waitKey(20) == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
## 키보드
img = cv2.imread('./fig/fig/puppy.bmp', 0)
cv2.imshow('img', img)
img1 = img.copy()

while True :
    key = cv2.waitKey()
    
    if key == ord('i') :        # i key를 누르면 gray scale
        img = 255 - img
        cv2.imshow('img', img)
        
    elif key == ord('e') :  # e key 를 누르면 edge 스케일
        img = cv2.Canny(img, 50, 150)
        cv2.imshow('img', img)
        
    elif key == ord('r') : # r키를 누르면 원래의 사진으로
        img = img1.copy()
        cv2.imshow('img', img)
        
    elif key == ord('s') : # s 키를 누르면 저장
        cv2.imwrite('output.png', img)
        
    elif key == 27 :
        break

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
def call_mouse(event, x, y, flags, param) : # x,y의 좌표값을 반환, 마우스 클릭할 시 좌표가 표시됨
    if event == cv2.EVENT_LBUTTONDOWN : 
        print('left button down :', x, y)
        
    elif event == cv2.EVENT_LBUTTONUP: 
        print('left button up :',x, y)

    elif event == cv2.EVENT_MOUSEMOVE :
        if flags == cv2.EVENT_FLAG_CTRLKEY : # CTRL 키를 누르면 마우스 이동시 좌표 표시
            print('Mouse move :', x, y)

img = np.ones((480, 640, 3), np.uint8)*255

cv2.namedWindow('img')
# setMouseCallback(windowName, onMouse [, param]) -> None
cv2.setMouseCallback('img', call_mouse, img) # 어디다가 마우스 좌표를 받을거고, 그 창의 이름, 어디다가 쓸건지


cv2.imshow('img', img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
def call_mouse(event, x, y, flags, param):
    global oldx, oldy
    
    
    if event == cv2.EVENT_LBUTTONDOWN:
        oldx, oldy = x, y
        
    
#     elif event == cv2.EVENT_LBUTTONUP:
#         print('Left button up :', x,y)
        
#         line(img, pt1, pt2, color[, thickness[, lineType[, shift]]]) -> img
    elif event == cv2.EVENT_MOUSEMOVE:
        if flags == cv2.EVENT_FLAG_LBUTTON:
            cv2.line(img, (oldx, oldy), (x, y), (0, 0, 255), 5, cv2.LINE_AA)
            cv2.imshow('image', img)
            oldx, oldy = x, y
            

img = np.ones((480, 640, 3), np.uint8)*255

cv2.namedWindow('image')
cv2.setMouseCallback('image', call_mouse, img)


cv2.imshow('image', img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print('camera open failed')
    sys.exit()
    
f_flag = False
i_flag = False


while True :
    
    ret, frame = cap.read()
    
    if not ret : 
        print('frame read failed')
        break
        
    if f_flag :
        frame = cv2.flip(frame,1)
        
    if i_flag:
        frame = cv2.Canny(frame, 50, 150)
    
    cv2.imshow('img', frame)
    
    key = cv2.waitKey(20)
    
    if key == 27:
        break
        
    elif key == ord('f') : 
        f_flag = not f_flag
    elif key == ord('i'):
        i_flag = not i_flag
        
cap.release()
cv2.destroyAllWindows()
```

```python
def call_trackbar(pos) : 
    img[:] = pos
    cv2.imshow('img', img)

img = np.ones((400,600), np.uint8)


cv2.imshow('img', img)

cv2.createTrackbar('Level', 'img', 0, 255, call_trackbar)

cv2.waitKey()
cv2.destroyAllWindows()
```
