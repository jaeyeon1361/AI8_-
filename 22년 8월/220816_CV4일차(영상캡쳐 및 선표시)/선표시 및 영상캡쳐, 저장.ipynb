```python
img_files = glob.glob('./fig/my_images/*.*')
idx=0
cv2.namedWindow('scene', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

while True : 
    img = cv2.imread(img_files[idx])
    
    if img is None :
        print('image read failed')
        break
        
    cv2.imshow('scene', img)
    
    if cv2.waitKey(3000) == 27:
        break
        
    idx += 1
    
    if idx >= len(img_files) :
        idx = 0
        
cv2.destroyAllWindows()
```

```python
img = np.full((600, 1200,3), (255, 255,255), np.uint8)


## 직선귿기
## line(img, pt1, pt2, color[, thickness[, lineType[, shift]]]) -> img
# img : 그림을 그릴 영상
# pt1, pt2 : 직선의 시작점과 끝점
# color : 직선의 색
# thinkness : 선두께, 기본은 1
# linetype
cv2.line(img, (100,100),(300,100), (0,0,255), 5, cv2.LINE_8)
cv2.line(img,(300, 100), (200, 300), (0,0,255), 10, cv2.LINE_4)
# 방향표시
cv2.arrowedLine(img, (400, 100), (400,300), (255,0,0), 10, cv2.LINE_AA)

# 네모
# rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]]) -> im
cv2.rectangle(img, (100,400), (400,500), (0,0,255), 5)
cv2.rectangle(img, (120,420,260,60),(0,255,0), -1 )


cv2.circle(img, (700,300), 100, (0,0,255), 10, cv2.LINE_AA)
# ellipse(img, center, axes, angle, startAngle, endAngle, color[, thickness[, lineType[, shift]]]) -> img
cv2.ellipse(img, (700,300),(50,100), 20, 0, 270, (0,255,255), 10, cv2.LINE_AA)

text = 'Open cv' + cv2.__version__
# putText(img, text, org, fontFace, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]]) -> img
cv2.putText(img, text, (700,500), cv2.FONT_HERSHEY_COMPLEX, 1, (128,128,128), 1, cv2.LINE_AA)

cv2.imshow('image', img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cap  = cv2.VideoCapture(0)

if not cap.isOpened() :  # -> 캠 존재여부 확인 True 나오면 됨
    print('camera open failed')
    sys.exit()
    
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) # -> 캠이 나오는 창 가로세로 길이
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# fps/frame for seconed
fps = int(cap.get(cv2.CAP_PROP_FPS)*0.7)
# fourcc/동영상 저장방식
fourcc = cv2.VideoWriter_fourcc(*'DIVX')

out = cv2.VideoWriter('edge_cvt.avi', fourcc, fps, (w,h))
    
    

while True : 
    ret, frame = cap.read()
    
    if not ret : 
        print('frame read failed')
        break
#    white = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    edge = cv2.Canny(frame,20,150)
    edge_cvt = cv2.cvtColor(edge, cv2.COLOR_GRAY2BGR)
#    inverse = 255 - white


    out.write(edge_cvt) # -> 저장할 영상 정보
    
#    cv2.imshow('image', frame)
    cv2.imshow('edge_cvt', edge_cvt)
#    cv2.imshow('edge', edge)
#    cv2.imshow('white', white)
#    cv2.imshow('inverse', inverse)
    
    if cv2.waitKey(20) == 27 :
        break
        
cap.release()
cv2.destroyAllWindows()
```
