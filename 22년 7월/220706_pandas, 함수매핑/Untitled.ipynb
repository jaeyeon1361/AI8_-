```python
# 데이터프레임 객체에 함수 매핑 : 데이터프레임.pipe(매핑함수) 226p ~ 
# 각 열의 Nan 자료를 찾아서 df 반환
def missing_value(x): 
    return x.isnull()

# 각 열의 Nan 갯수 반환 : 데이터프레임-> 시리즈반환
def missing_count(x):
    return missing_value(x).sum()

# 데이터 프레임 Nan의 총 갯수 : 값을 반환
def total_number_missing(x) :
    return missing_count(x).sum()
```

```python
df_1 = df.pipe(missing_value)
df_1.head()
```

```python
sr_1 = df.pipe(missing_count)
sr_1
```

```python
value = df.pipe(total_number_missing)
value
```

```python
# 2단계 : 그룹연산을 그룹의 객체 적용
titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]

grouped = df.groupby(['class'])  # 1단계
grouped
```

```python
std_all = grouped.std()
std_all
```

```python
grouped['fare'].std()
```

```python
# 사용자 정의 함수를 그룹 객체에 적용 : 그룹객체.agg(매핑함수)
agg_max_min = grouped[['age', 'fare']].agg(lambda  x: max(x) - min(x))
agg_max_min
```

```python
# 동시에 여러개의 함수를 적용 : 그룹객체.agg([함수1, 함수2, ...])
agg_all = grouped.agg(['min', 'max', 'mean'])
agg_all
```

```python
# 컬럼마다 다른 함수를 적용 : 그룹객체.agg(['열' : 함수, '열2' : 함수2 ...])
agg_def = grouped.agg({'sex' : 'min', 'fare' : ['mean', 'std']})
```

```python
# 그룹연산데이터 변환 : group 객체.transform(매핑함수)
age_std = grouped['age'].std()
age_std
```

```python
age_mean = grouped['age'].mean()
age_mean
```
