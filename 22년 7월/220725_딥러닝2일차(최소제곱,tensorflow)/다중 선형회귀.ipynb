```python
# tensorflow를 이용한 선형회귀
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(1, input_dim= 1, activation = 'linear'))
model.compile(optimizer = 'sgd', loss = 'mse')
model.fit(x,y,epochs=2000)
plt.scatter(x,y)
plt.plot(x, model.predict(x), 'r')
plt.show()
```

```python
hour = 7
prediction = model.predict([hour])
print("%.f 시간을 공부 할 경우의 예상 점수는 %.02f점 입니다." % (hour, prediction))
```

```python
# tensorflow 다중선형회귀
x = np.array([[2,0], [4,4], [6,2], [8,3]])
y = np.array([81, 93, 91, 97])

model = Sequential()

# 입력 변수가 두개 (학습시간, 과외시간)이므로 input_dim에 2 입력
model.add(Dense(1, input_dim=2, activation='linear'))

model.fit(x, y, epochs=2000)
```

```python
hour=7
private_calss = 4
prediction = model.predict([[hour, private_calss]])

print("%.f 시간을 공부하고, %.f시간의 과외를 받는 경우, 예상 점수는 %.02f점 입니다." % (hour, private_calss, prediction))
```
