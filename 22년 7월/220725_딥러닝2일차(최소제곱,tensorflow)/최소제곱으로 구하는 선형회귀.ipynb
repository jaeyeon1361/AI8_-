```python
import numpy as np
```

```python
## 최소제곱
x = np.array([2,4,6,8])
y = np.array([81, 93, 91, 97])
# x= 평균 / y=성적
```

```python
mx = np.mean(x)
my = np.mean(y)
```

```python
# 기울기 공식의 분모 부분
divisor = sum([(i-mx)**2 for i in x])
```

```python
# 기울기 공식의 분자 부분
def top(x, mx, y, my):
    d = 0
    for i in range(len(x)):
        d += (x[i] - mx) * (y[i] - my) # 각 x 에서 평균을 빼고, 각 y에서 평균을 뺌
    return d
dividend = top(x, mx, y, my)
```

```python
print("분모:", divisor)
print("분자:", dividend)
```

```python
a = dividend / divisor
b = my - (mx*a)
```

```python
print("기울기 a = ", a)
print("y절편 b = ", b)
```

```python
## 평균 제곱 오차 _ 주어진 데이터에서 오차값을 낸다음에 그것을 제곱 
fake_a = 3
fake_b = 76
# 임의로 정한 x와 y

def predict(x):
    return fake_a * x + fake_b
```

```python
predict_result = []
```

```python
for i in range(len(x)) : 
    predict_result.append(predict(x[i]))
    print("공부시간=%.f, 실제점수 =%.f, 예측점수=%.f" % (x[i], y[i], predict(x[i])))
```

```python
# 평균제곱오차 파이썬으로
n = len(x)
def mse(y, y_pred) : 
    return (1/n) * sum((y-y_pred)**2)
print("평균 제곱 오차 :" + str(mse(y, predict_result)))
```

```python
# 선형회귀
y_pred = a * x + b
error = y = y_pred

a_diff = (2/n) * sum(-x * (error))
b_diff = (2/n) * sum(-(error))
```

```python
lr = 0.03 
a = a - lr * a_diff
b = b - lr * b_diff
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.show()
```

```python
    a = 0
    b = 0
lr = 0.03 # 학습률
epochs = 2001 # 몇번 반복할 것인지 설정
n = len(x) # X읭 값의 총 몇개인지 센다
```

```python
for i in range(epochs):
    y_pred = a * x + b
    error = y - y_pred # 실제 값과 비교한 오차
    a_diff = (2/n) * sum(-x * (error)) # 오차 함수를 a 로 편미분
    b_diff = (2/n) * sum(-(error)) # 오차 함수를 b로 편미분
    a = a - lr * a_diff # 학습률을 곱해 기존의 a 값을 업데이트
    b = b - lr * b_diff # 위와 동일하게 b 값을 업데이트
    
    if i % 100 == 0 :
        print("epoch =%.f, 기울기=%.04f, 절편=%.04f" % (i, a, b))
        plt.scatter(x,y)
        plt.plot(x, y_pred, 'r')
        plt.show()
```
