```python
import pandas as pd
import matplotlib.pyplot as plt # 그래프 모듈 가져옴
```

```python
df = pd.read_excel('dataset/data/시도별 전출입 인구수.xlsx', header=0)
df.head()
```

```python
# 위에 있는 Nan 을 이전 자료로 대체 : df.gillna(mathod='ffoll')
df = df.fillna(method='ffill')
df.head()
```

```python
# 전출지가 '서울특별시'에서 다른 도시로 이전한 인구수의 분포를 분석하고자 함
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] !='서울특별시')
mask[10:40]
```

```python
df_seoul = df[mask] # mask 의 값이 True인 값만 추출
df_seoul
```

```python
# 위에서 전출지별의 값은 모두 '서울특별시' 임으로 컬럼 삭제 진행
df_seoul = df_seoul.drop('전출지별', axis=1)
df_seoul
```

```python
# 전입지별 컬럼명을 '전입지'로 변경
df_seoul.rename({'전입지별' : '전입자'}, axis=1, inplace=True)
df_seoul
```

```python
# '전입지' 칼럼을 인덱스로 설정
df_seoul.set_index('전입자', inplace=True)
```

```python
# df_seoul에서 전입자가 경기도인 자료만 추출
sr_one = df_seoul.loc['경기도', :]
sr_one
```

```python
# 선그래프 그리기
plt.plot(sr_one)
```

```python
plt.plot(sr_one.index, sr_one.values)
```

```python
# 폰트문제 해결
from matplotlib import rc

rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)
```

```python
# 그래프에 제목 추가, 축 이름 추가
# 서울 -> 경기도 이동인구
# x축 이름 -> '기간' , y축 이름 -> '이동 인구 수'
plt.title('서울 -> 경기도 이동 인구')
plt.xlabel('기간')
plt.ylabel('이동 인구 수')

plt.show
```

```python
plt.figure(figsize=(14,5)) # 그래프 사이즈 지정

# x축 눈금 라벨 회전 : plt.xticks(rotation = 'vertical')
plt.xticks(rotation = 'vertical')
plt.plot(sr_one)

plt.title('서울 -> 경기도 이동 인구')
plt.xlabel('기간')
plt.ylabel('이동 인구 수')
plt.legend(labels=['서울->경기'], loc = 'best')

plt.show()
```

```python
fig = plt.figure(figsize= (14,10))
ax1 = fig.add_subplot(2, 1, 1) # 한 행에 2개, 1열에 1개, 위치
ax2 = fig.add_subplot(2, 1, 2) 

ax1.plot(sr_one, marker ='o')
ax2.plot(sr_one, color = 'olive', label = '서울->경기') # 색변경
ax2.legend(loc='best')


ax1.set_title('서울 -> 경기 이동 인구')
ax1.set_xticklabels(sr_one.index, rotation=75)
ax2.set_xticklabels(sr_one.index, rotation=75)
plt.show()
```

```python
# df_2 서울에서 충청남도 경상북도 강원도 전라남도로 이동한 자료를 가져옴
# 1980년도에서 2016년까지의 자료만
col_years =list(map(str, range(1980, 2017))) # 저장된 자료가 str, int-> str로 변경
df_2 =df_seoul.loc[['충청남도', '경상북도', '강원도', '전라남도'], col_years]
df_2
```

```python
# 그래프 : 각 도시를 하나의 그래프에, 한페이지 4개의 그래프로
fig = plt.figure(figsize=(15, 15))

# 충청남도 그래프 지정
ax1 = fig.add_subplot(2, 2, 1) # 충남
ax2 = fig.add_subplot(2, 2, 2) # 경북
ax3 = fig.add_subplot(2, 2, 3) # 강원도
ax4 = fig.add_subplot(2, 2, 4) # 전남

ax1.plot(col_years, df_2.loc['충청남도', : ], marker='o', color='blue', 
        label = '서울 -> 충남')
ax2.plot(col_years, df_2.loc['경상북도', : ], marker='o', color='green', 
        label = '서울 -> 경북')
ax3.plot(col_years, df_2.loc['강원도', : ], marker='o', color='red', 
        label = '서울 -> 강원도')
ax4.plot(col_years, df_2.loc['전라남도', : ], marker='o', color='skyblue', 
        label = '서울 -> 전남')

# 범례표시
ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()

# 각 그래프의 제목 추가
ax1.set_title('서울->충남 인구이동', size = 12 )
ax2.set_title('서울->경북 인구이동', size = 12 )
ax3.set_title('서울->강원도 인구이동', size = 12 )
ax4.set_title('서울->전남 인구이동', size = 12 )

# 각 그래프의 xti
ax1.set_xticks(ticks =col_years, labels=col_years, rotation=75)
ax2.set_xticks(ticks =col_years, labels=col_years, rotation=75)
ax3.set_xticks(ticks =col_years, labels=col_years, rotation=75)
ax4.set_xticks(ticks =col_years, labels=col_years, rotation=75)

plt.show()
```

```python
# twin 그래프 : x축은 동일 y축의 범위가 서로 다른 그래프 그리기
df = pd.read_excel('./dataset/data/남북한발전전력량.xlsx')
df.info()
```

```python
df = df.loc[5:] # 북한의 수력, 화력 발전량의 추이를 분석
df
```

```python
df = df.drop('전력량 (억㎾h)', axis=1) # 전력량 없앰
df
```

```python
# 발전 전력별 컬럼을 인덱스로
df.set_index('발전 전력별', inplace=True)
df
```

```python
df_t = df.T
df_t
```

```python
# 합계 -> 총 발전량으로 변경
df_t = df_t.rename(columns={'합계' : '총발전량'})
df_t 
```

```python
# 증감율 : (총발전량 / (총발전량 - 전년도 총 발전량) -1 )* 100
# ['총발전량' - 1년]
print(df_t['총발전량'], df_t['총발전량'].shift(1))
```

```python
df_t['전년도 발전량'] = df_t['총발전량'].shift(1)
df_t
```

```python
# 증감율 : (총발전량/전년도 발전량 -1) * 100
df_t['증감율'] = (df_t['총발전량']/df_t['전년도 발전량'] -1) * 100
df_t
```

```python
# 2개축 그래프 그리기 : 수력 화력 -> 막대 그래프, 증감율 -

ax1 = df_t[['수력', '화력']].plot(kind='bar', stacked=True)
ax2 = ax1.twinx()        # y축의 범위가 다른경우
ax2.plot(df_t.index, df_t.증감율, ls = '--', marker='o', label='전년대비 증감율(%)')

ax1.set_ylim(0, 500)
ax2.set_ylim(-50, 50)

ax1.set_xlabel("연도")
ax2.set_ylabel('발전량 (억 Kwh)')

plt.title("북한 전력 발전량(1990 ~ 2016)")
plt.legend(loc='best')
plt.show()
```
